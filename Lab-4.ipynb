```python
from sklearn.datasets import load_digits, load_wine, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python

# Загрузка данных
digits_data = load_digits()
wine_data = load_wine()
cancer_data = load_breast_cancer()

# Изучение данных
print("Digits dataset:")
print(digits_data.DESCR)
print("Wine dataset:")
print(wine_data.DESCR)
print("Breast Cancer dataset:")
print(cancer_data.DESCR)
```

```python
# Разделение данных на обучающий и тестовый наборы (пример для digits dataset)
X_digits, y_digits = digits_data.data, digits_data.target
X_digits_train, X_digits_test, y_digits_train, y_digits_test = train_test_split(X_digits, y_digits, test_size=0.2, random_state=42)
```

```python
# Визуализация примера из digits dataset
plt.imshow(X_digits[0].reshape(8, 8), cmap='gray')
plt.title(f"Label: {y_digits[0]}")
plt.show()
```

```python
# Построение модели k-ближайших средних
knn = KNeighborsClassifier(n_neighbors=5)  # Пример с 5 ближайшими соседями
knn.fit(X_digits_train, y_digits_train)

# Протестировать модель на произвольном наборе данных (пример для digits dataset)
random_sample = X_digits_test[0].reshape(1, -1)
prediction = knn.predict(random_sample)
print(f"Predicted label: {prediction[0]}")
```

```python
# Оценка качества модели (пример для digits dataset)
y_digits_pred = knn.predict(X_digits_test)
accuracy = accuracy_score(y_digits_test, y_digits_pred)
report = classification_report(y_digits_test, y_digits_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", report)
```
