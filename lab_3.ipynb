```python
#Задача - 2

import matplotlib.pyplot as plt
import numpy as np

# Создаем данные для графика квадратичной функции
x = np.linspace(-5, 5, 100)  # Задаем диапазон x
y = x**2  # Квадратичная функция

# Строим график
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График квадратичной функции')
plt.grid(True)

# Отображаем график
plt.show()
```

```python
# Задача - 3 

# Создаем данные для графика гармонической функции
x = np.linspace(0, 2 * np.pi, 100)  # Задаем диапазон x
y = np.sin(x)  # Гармоническая функция (синус)

# Строим график
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График гармонической функции')
plt.grid(True)

# Отображаем график
plt.show()
```

```python
#Задача - 4

# Создаем данные для обоих графиков
x1 = np.linspace(-5, 5, 100)
y1 = x1**2
x2 = np.linspace(0, 2 * np.pi, 100)
y2 = np.sin(x2)

# Разбиваем окно на два поля
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, первое поле
plt.plot(x1, y1)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График квадратичной функции')
plt.grid(True)

plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, второе поле
plt.plot(x2, y2)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('График гармонической функции')
plt.grid(True)

# Отображаем графики
plt.tight_layout()
plt.show()
```

```python
#Задача - 5

# Создаем данные для графиков
x1 = np.linspace(-5, 5, 100)
y1 = x1**2
x2 = np.linspace(0, 2 * np.pi, 100)
y2 = np.sin(x2)

# Строим графики и добавляем легенду
plt.plot(x1, y1, label='Квадратичная функция')
plt.plot(x2, y2, label='Гармоническая функция')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Графики квадратичной и гармонической функций')
plt.legend()
plt.grid(True)

# Отображаем графики
plt.show()
```

```python
# задача - 6

import matplotlib.pyplot as plt
import numpy as np

# Создаем параметрические данные
t = np.linspace(0, 10, 100)  # Задаем диапазон параметра t
x = np.cos(t)  # x(t)
y = np.sin(t)  # y(t)

# Строим график параметрической функции
plt.plot(x, y)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('График параметрической функции')
plt.grid(True)

# Отображаем график
plt.show()
```

```python
# Задача - 7

import matplotlib.pyplot as plt

# Годы и данные по прожиточному минимуму
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
minimum_wage = [
    4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 
    4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32
]

# Создаем столбиковую диаграмму
plt.figure(figsize=(12, 6))  # Размер графика

plt.bar(years, minimum_wage, color='skyblue')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.xticks(years, rotation=45)  # Поворачиваем названия годов для удобства

plt.tight_layout()
plt.show()
```

```python
# Задача - 8

import matplotlib.pyplot as plt

# Наименования годов и данные по прожиточному минимуму
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
minimum_wage = [
    4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 
    4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32
]

# Создаем круговой график
plt.figure(figsize=(8, 8))  # Размер графика

# Строим круговой график
plt.pie(minimum_wage, labels=years, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Чтобы круг был кругом

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')

plt.show()
```

```python
# Задача - 9

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Создаем данные для каркасной поверхности
u = np.linspace(-5, 5, 100)
v = np.linspace(-5, 5, 100)
u, v = np.meshgrid(u, v)
x = u
y = v
z = u**2 + v**2  # Пример каркасной поверхности (параболоид)

# Создаем трехмерное окно
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Строим каркасную поверхность
ax.plot_wireframe(x, y, z, color='b', linewidth=0.5)

# Настройки подписей
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')
ax.set_title('Каркасная поверхность')

# Отображаем график
plt.show()
```
